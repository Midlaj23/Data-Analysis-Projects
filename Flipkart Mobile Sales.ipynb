```python
sns.distplot(df['sales_price'])
```

##### This skewed distribution suggests that the majority of products are relatively affordable, with fewer products at higher price points.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='discount_percent', y='sales', data=df)
plt.title('Impact of Discounts on Sales')
plt.xlabel('Discount Percentage')
plt.ylabel('Sales')
plt.xticks(rotation = 90)
plt.show()
```

 - As the discount percentage increases, sales tend to decrease
 - The majority of sales occur at lower discount percentages
   - moderate discounts appear to be most effective for driving mobile sales on Flipkart.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ratings'], bins=5, kde=False, color='skyblue')
plt.title('Distribution of Product Ratings')
plt.xlabel('Product Ratings')
plt.ylabel('Frequency')
plt.show()
```

##### The majority of product ratings fall within the higher rating range, indicating that customers are generally satisfied with their purchases

```python
high_rated_mobiles = df[df['ratings']==df['ratings'].max()]
print("Products with the Highest Ratings:")
print(high_rated_mobiles[['brand', 'model', 'ratings']])
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='brand', data=high_rated_mobiles, palette='viridis')
plt.title('Distribution of Products with Highest Ratings')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
```

- The products with the highest ratings (4.6) include several models from Apple, such as iPhone XR, iPhone 12

```python
low_rated_mobiles = df[df['ratings'] == df['ratings'].min()]
print("Products with the Lowest Ratings:")
print(low_rated_mobiles[['brand', 'model', 'ratings']])
```

- The product with the lowest rating in the dataset is the Samsung Galaxy Fold 2, which has a rating of 3.0

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='brand', y='discount_percent', data=df, palette='rainbow')
plt.title('Distribution of Discount Percentages Across Brands')
plt.xlabel('Brand')
plt.ylabel('Discount Percentage')
plt.show()
```

- Apple has the lowest discount range.
- Poco, Realme, and Samsung have similar median discount percentages but vary in their ranges.
   - Xiaomi stands out with its diverse discount offerings, while Apple remains conservative in its discounts

```python
plt.figure(figsize=(14, 6))
sns.barplot(x='RAM', y='sales', data=df, palette='Set1')
plt.title('Relationship between RAM and Sales')
plt.xlabel('RAM')
plt.ylabel('Sales')
plt.show()

plt.figure(figsize=(14, 6))
sns.barrplot(x='ROM', y='sales', data=df, palette='Set2')
plt.title('Relationship between ROM and Sales')
plt.xlabel('ROM')
plt.ylabel('Sales')
plt.show()
```

- Phones with 2GB RAM have the highest sales percentage, nearly reaching 100
- Other RAM sizes (3GB, 4GB, 6GB, 8GB, 12GB) show significantly lower sales percentages
- Phones with 16GB ROM have the highest sales percentage.
- Thereâ€™s a significant drop in sales after 16GB ROM.

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='ratings', y='sales', data=df, palette='viridis')
plt.title('Relationship between Ratings and Sales')
plt.xlabel('Ratings')
plt.ylabel('Sales')
plt.show()
```

- Products with ratings around 4 and above show a positive correlation with sales.
- As ratings increase, there is a noticeable upward trend in sales.
- The left-skewed pattern indicates that products with lower ratings have varying sales, while higher ratings are consistently associated with higher sales

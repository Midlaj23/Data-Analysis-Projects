```python
top_selling_products = df.groupby('Product Name')['Sale Price'].sum().nlargest(10)

product_performance = df.groupby('Product Name')[['Star Rating', 'Number Of Ratings', 'Number Of Reviews']].mean()
product_performance = product_performance.loc[top_selling_products.index]
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=top_selling_products.index, y=top_selling_products.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Product Name')
plt.ylabel('Total Sale Price')
plt.title('Top 10 Selling Products')
plt.tight_layout()
plt.show()
```

 - APPLE iPhone 12 Pro (Pacific Blue - 512 GB) is the top-selling product, with the highest total sale price among the top 10 selling products.
 - Following that, other iPhone models like "APPLE iPhone 12 Pro (Silver - 512 GB)" and "APPLE iPhone 11 Pro Max (Gold - 256 GB)" also have substantial total sale prices.
 - The plot provides insights into the popularity and revenue generated by each of the top-selling iPhone models.


```python
highest_rated = df.sort_values(by="Star Rating", ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=highest_rated["Product Name"], y=highest_rated["Number Of Ratings"], palette='inferno')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Product Name')
plt.ylabel('Number of Ratings')
plt.title('Number of Ratings of Highest Rated iPhones')
plt.tight_layout()
plt.show()
```

 - The iPhone with the highest number of ratings among the top 10 highest-rated iPhones is "APPLE iPhone 12 (White - 128 GB)" with over 3,000 ratings.
 - The iPhone with the lowest number of ratings among the top 10 is "APPLE iPhone 12 Pro (Blue - 128 GB)" with around 500 ratings.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Number Of Ratings", y="Sale Price", size="Discount Percentage", sizes=(20, 200))
plt.title("Relationship between Sale Price and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Sale Price")
plt.show()
```

 - iPhones with lower sale prices tend to have a wider range of ratings, while higher-priced iPhones generally have fewer ratings.
 -  iPhones with lower sale prices are sold more in India.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Number Of Ratings", y="Discount Percentage", size="Sale Price", sizes=(20, 200))
plt.title("Relationship between Discount Percentage and Number of Ratings of iPhones")
plt.xlabel("Number Of Ratings")
plt.ylabel("Discount Percentage")
plt.show()
```

 - There is a linear relationship between the discount percentage on iPhones on Flipkart and the number of ratings. It means iPhones with high discounts are sold more in India.

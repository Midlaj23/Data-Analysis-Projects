```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/amankharwal/Website-data/master/Virat_Kohli.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Runs'].mean()
```

-  Batting average of approximately 46.85 runs per match in ODI cricket can be considered quite good. In cricket

```python
matches = df.index
plt.figure(figsize=(10, 6))
plt.plot(matches, df['Runs'], marker='o', linestyle='-',color = 'r')
plt.xlabel('Match Number')
plt.ylabel('Runs Scored')
plt.title('Virat Kohli ODI Performance Over Time')
plt.grid(True)
plt.show()
```

-  Virat Kohliâ€™s ODI performance has been a mix of brilliance, consistency, and occasional challenges. His ability to score centuries and anchor innings makes him a vital asset for the Indian cricket team

```python
sns.boxplot(x='Pos', y='Runs', data=df)
plt.xlabel('Batting Position')
plt.ylabel('Runs Scored')
plt.title('Impact of Batting Position on Runs Scored')
plt.show()
```

 - 3rd position tend to score more runs compared to other positions.
 - This suggests that the 3rd position is crucial for a teamâ€™s scoring

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Pos', y='4s', data=df)
plt.xlabel('Batting Position')
plt.ylabel('4s')
plt.title('4s Scored by Batting Position')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pos', y='6s', data=df)
plt.xlabel('Batting Position')
plt.ylabel('6s')
plt.title('6s Scored by Batting Position')
plt.grid(True)
plt.show()
```

- Virat Kohliâ€™s ability to find the boundary is most pronounced when he bats at the 4th position

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Opposition', y='Runs', data=df, palette='plasma')
plt.title('Most Runs Against Teams')
plt.xlabel('Opposition')
plt.ylabel('Runs')
plt.xticks(rotation=45) 
plt.show()
```

- Virat Kohli performs exceptionally well against Bangladesh and New Zealand, scoring the highest runs, followed closely by strong performances against West Indies and Australia, while his runs against England and South Africa are slightly lower.

```python
ground_performance = df.groupby('Ground')['Runs'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
ground_performance.plot(kind='bar', color='r')
plt.xlabel('Ground')
plt.ylabel('Average Runs Scored')
plt.title('Virat Kohli Average Runs Scored at Different Grounds (Descending Order)')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

- Virat Kohli has scored the highest average runs at Fatullah, Napier, and Canberra cricket grounds, while his lowest average runs were at Bulawayo.

```python
centuries = df[df['Runs'] >= 100]
plt.figure(figsize=(10, 6))
sns.barplot(x='Inns', y='Runs', data=centuries, palette='inferno')
plt.title('Centuries By Virat Kohli in First Innings Vs. Second Innings')
plt.xlabel('Innings')
plt.ylabel('Runs')
plt.show()
```

- Virat Kohli tends to perform exceptionally well in the second innings of cricket matches

```python
high_sr_innings = df[df['SR'] > 120]
plt.figure(figsize=(8, 6))
sns.countplot(x='Inns', data=high_sr_innings, palette='Set2')
plt.title('Innings with Strike Rate > 120 by Virat Kohli')
plt.xlabel('Innings')
plt.ylabel('Frequency')
plt.show()
```

- Kohli has played more innings with a high strike rate in the first innings compared to the second innings.

```python
dismissal_counts = df["Dismissal"].value_counts()
labels = dismissal_counts.index
counts = dismissal_counts.values

plt.figure(figsize=(8, 8))
plt.pie(counts, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Dismissals of Virat Kohli')
plt.show()
```

 - The most common dismissal type for Kohli is being caught (63.6%)

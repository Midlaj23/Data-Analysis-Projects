```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Documents\15 days project datasets\companies_by_revenue.csv")
df
```

```python
df.info()
```

```python
df['Revenue (USD millions)'] = df['Revenue (USD millions)'].str.replace(',', '').astype(float)
df['Revenue (USD millions)']
```

```python
revenue_stats = df['Revenue (USD millions)'].describe()
revenue_stats
```

 - The average revenue of these companies is approximately 117,566 million.
 - The lowest recorded revenue is 44,200 million.
 - The highest recorded revenue is 611,289 million.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Revenue (USD millions)'], kde=True)
plt.title('Distribution of Revenue')
plt.xlabel('Revenue (USD millions)')
plt.ylabel('Frequency')
plt.show()
```

 - Most companies fall within the lower range of revenue.
 - Over 40 companies have revenues around 100,000 USD millions.
 - Fewer companies achieve higher revenues

```python
df['Revenue growth'] = df['Revenue growth'].str.rstrip('%').astype(float)
```

```python
headquarters_counts = df['Headquarters'].value_counts().head(10)

plt.figure(figsize=(12, 8))
headquarters_counts.plot(kind='bar')
plt.title('Top 10 Most Common Headquarters Locations')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

 - New York City, New York has the highest frequency among the top 10 headquarters locations, with 12 companies.
 - Houston, Texas follows with 5 companies having their headquarters located there
 - Atlanta, Georgia, and Cincinnati, Ohio have 3 and 2 companies

```python
industry_counts = df['Industry'].value_counts().head(10)

plt.figure(figsize=(10, 6))
industry_counts.plot(kind='bar', color='red')
plt.title('Top 10 Industries Represented')
plt.xlabel('Industry')
plt.ylabel('Number of Companies')
plt.xticks(rotation=45)
plt.show()
```

 - Financials, Retail, and Petroleum industry are the most frequent industries, each represented by 11 companies.

```python
df['Employees'] = df['Employees'].apply(lambda x: ''.join(char for char in x if char.isdigit()))
df['Employees'] = df['Employees'].astype(int)
```

```python
employee_stats = df['Employees'].describe()
employee_stats
```

 - On average, there are approximately 168,427 employees in the companies.
 - The minimum number of employees in a company is 3,616, while the maximum is 2,100,000.
 - The median value suggests that half of the companies have fewer than 89,638 employees, while the other half have more.

```python
industry_employees = df.groupby('Industry')['Employees'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
industry_employees.plot(kind='bar', color='indigo')
plt.title('Top 10 Industries by Total Number of Employees')
plt.xlabel('Industry')
plt.ylabel('Total Employees')
plt.xticks(rotation=45)
plt.show()
```

 -  The retail sector leads with the highest number of employees, followed by industries like healthcare, financials, and transportation

```python
top_industries_revenue = df.groupby('Industry')['Revenue (USD millions)'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 8))
top_industries_revenue.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('dark'), startangle=140)
plt.title('Top 10 Industries by Total Revenue')
plt.ylabel('')
plt.show()
```

 - Retail stands out as the highest revenue earner, followed closely by the petroleum industry, healthcare, and financial sectors.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Documents\15 days project datasets\netflix_titles.csv")
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['director'].fillna(' ',inplace = True)
df['cast'].fillna(' ',inplace = True)
df['country'].fillna(' ',inplace = True)
df['date_added'].fillna(' ',inplace = True)
df['rating'].fillna(' ',inplace = True)
```

```python
df
```

```python
plt.figure(figsize=(8, 8))
type_counts = df['type'].value_counts()
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Content Types')
plt.show()
```

- The Netflix dataset contains a larger proportion of movies compared to TV shows, suggesting that movies are more prevalent on the platform.

```python
top_10_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10, 6))
sns.countplot(x='country', data=df[df['country'].isin(top_10_country.index)], palette='viridis')
plt.title('Distribution of Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

 - The dataset contains the highest number of shows and movies from the United States

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df, palette='Dark2', order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

- The most common ratings in the dataset are TV-MA, TV-14, and TV-PG

```python
directors_without_null = df[df['director'] != '']

top_directors = directors_without_null['director'].value_counts().iloc[1:].head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.index, y=top_directors.values, palette='rainbow')
plt.title('Top 10 Directors on Netflix ')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

 - Top directors on Netflix include Rajiv Chilaka, Raúl Campos, Jan Suter, Suhas Kadav, and Marcus Raboy.

```python
listed_in_without_null = df[df['listed_in'] != '']

top_listed_in = listed_in_without_null['listed_in'].value_counts().iloc[1:].head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_listed_in.index, y=top_listed_in.values, palette='muted')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

 - The top genres on Netflix include Documentaries, Stand-Up Comedy, Comedies, Dramas, International Movies, and Kids' TV. These genres are the most prevalent among the available content on the platform,

```python
plt.figure(figsize=(10, 6))
top_cast = df[df['cast'] != ' ']['cast'].value_counts().head(10)
sns.barplot(x=top_cast.index, y=top_cast.values, palette='inferno')
plt.title('Top 10 Cast Members on Netflix')
plt.xlabel('Cast Member')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

 - The most popular cast members on Netflix feature well-known figures like David Attenborough and Samuel West

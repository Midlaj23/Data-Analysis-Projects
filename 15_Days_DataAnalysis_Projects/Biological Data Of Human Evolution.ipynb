```python
df['Genus_&_Specie'].value_counts().sort_values(ascending = False)
```

- The dataset contains an equal number of entries for each genus and species, with 500 occurrences for each

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Location', data=df,palette = 'inferno')
plt.title('Distribution of Location')
plt.xlabel('Location')
plt.ylabel('Frequency')
plt.show()
```

 - The dataset predominantly focuses on African locations, providing valuable insights into human evolution in that region.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Habitat'], kde=False)
plt.title('Distribution of Habitat')
plt.xlabel('Habitat')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

 - Savannah: Dominates with around 4000 occurrences, making it the most common habitat.
 - Mixed: Follows with around 3000 occurrences.
 - Forest/Savanna, and Peninsular habitats are less frequent, each with fewer than 1000 occurrences.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Zone'], color = 'r')
plt.title('Zone-wise Distribution of Species')
plt.xlabel('Zone')
plt.ylabel('Number of Species')
plt.show()
```

 - This dataset highlights the significant variation in species distribution across different zones, with the oriental zone exhibiting the highest biodiversity

```python
top_10_species = df.groupby('Genus_&_Specie')['Cranial_Capacity'].mean().nlargest(10)
plt.figure(figsize=(10, 6))
top_10_species.plot(kind='bar', color='r')
plt.title('Top 10 Species with Highest Cranial Capacity')
plt.xlabel('Species')
plt.ylabel('Cranial Capacity')
plt.xticks(rotation=90)
plt.show()
```

- In the top 10 species with the highest cranial capacity, "Homo Heidelbergensis" exhibits the highest average cranial capacity, with a value of approximately 1299.88 cubic centimeters. Following closely is "Homo Sapiens" with an average cranial capacity of around 1235.94 cubic centimeters

### Morphological Analysis:

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Time', y='Cranial_Capacity', data=df,color = 'r')
plt.title('Cranial Capacity Over Time')
plt.xlabel('Time Period')
plt.ylabel('Cranial Capacity')
plt.xticks(rotation=45)
plt.show()
```

- The scatter plot indeed shows a decrease in cranial capacity over time, rather than an increase. The peak occurs around the 1st time period, followed by a slight decline and subsequent stabilization. This pattern provides valuable insights into human evolution and the changes in cranial capacity across different eras.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Incisor_Size', data=df)
plt.title('Distribution of Incisor Size')
plt.xlabel('Incisor Size')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Behavioral and Ecological Analysis:

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Diet', data=df,palette = 'Set2')
plt.title('Dietary Habits')
plt.xlabel('Diet')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

- Omnivore: The most common diet, with a significantly higher frequency.
- Dry Fruits: Also popular, but less frequent than omnivores.
- Hard Fruits: Consumed by fewer individuals.
- Carnivorous: A less common diet choice.
- Soft Fruits: The least preferred diet.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['biped'],color = 'r')
plt.title('Bipedal Locomotion')
plt.xlabel('Bipedal Locomotion')
plt.ylabel('Frequency')
plt.show()
```

 - The most common category, represented by the tallest bar, indicates that modern forms of bipedal locomotion are prevalent

### Anatomical Analysis:

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Arms', data=df)
plt.title('Arms Morphology')
plt.xlabel('Arms Morphology')
plt.ylabel('Frequency')
plt.show()
```

 - Climbing: The most common arm type, indicated by the tallest bar, suggests that climbing arms are prevalent.

```python
foot_counts = df['Foots'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=foot_counts.index, y=foot_counts.values, palette='Set2')
plt.title('Distribution of Hominid Species by Foot Characteristics')
plt.xlabel('Foot Characteristics')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

 - The most common foot characteristic, represented by the tallest bar, indicates that many hominid species are adapted for walking

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Hip', data=df,palette = 'inferno')
plt.title('Hip Size')
plt.xlabel('Hip Size')
plt.ylabel('Frequency')
plt.show()
```

 - The most common hip size observed in the dataset is categorized as “wide.” These individuals likely have broader hips.
 - The least common category is “very modern,” representing individuals with a distinct hip size

```python
sexual_dimorphism_counts = df['Sexual_Dimorphism'].value_counts()
plt.figure(figsize=(10, 6))
sns.countplot(x='Sexual_Dimorphism', data=df, order=sexual_dimorphism_counts.index, palette='rainbow')
plt.title('Distribution of Sexual Dimorphism Among Hominid Species')
plt.xlabel('Sexual Dimorphism')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

 - High Sexual Dimorphism: A significant difference in body size between males and females is common in many hominid species. These species exhibit a clear distinction in physical proportions.
 - Medium/High Sexual Dimorphism: Some hominids fall into this category, indicating a moderate to substantial difference in body size between genders.
 - Reduced Sexual Dimorphism: Only a few hominid species show minimal differences in body size between males and females. Their physical proportions are more similar.

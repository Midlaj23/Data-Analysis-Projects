```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df  = pd.read_csv(r"C:\Users\Lenovo\Documents\15 days project datasets\dataset_salary_cyber.csv")
df
```

```python
df = df.drop(columns = ['salary_currency','remote_ratio','salary'])
```

```python
df[df['salary_in_usd'] == df['salary_in_usd'].max()]
```

- Senior level Detection engineer is the high salary job in Cyber security Field (Full time)

```python
df[df['salary_in_usd'] == df['salary_in_usd'].min()]
```

 - Entry Level Penetration tester and application Security engineer is the low salary jobs in Cyber Security Fields

```python
plt.figure(figsize = (10,6))
sns.histplot(df['salary_in_usd'],kde = True,bins = 20,color = 'r')
plt.title('Salary Distribution on Cyber Security Jobs')
plt.show()
```

 - Most of them Earn Up to 200000
 - Few of them Earn higher salaries in Cyber Security Fields

```python
top_10_salaries = df.sort_values(by='salary_in_usd', ascending=False).head(10)
top_10_salaries = top_10_salaries[['job_title', 'salary_in_usd']]
top_10_salaries
```

 - These are Top 10 High salary jobs in Cyber Security Fields

```python
plt.figure(figsize = (10,6))
sns.barplot(data = df,x='experience_level',y='salary_in_usd',palette = 'inferno')
plt.title('Salary Distribution on Experience level')
plt.show()
```

 - Executive level get High salary than Senior level
 - Entry level get the lowest salary 

```python
plt.figure(figsize = (10,6))
sns.barplot(data = df,x='company_size',y='salary_in_usd',palette = 'rainbow')
plt.title('Salary Distribution on Company Size')
plt.show()
```

 - medium level companies gives more salary than large companies
 - As usual Small Companies gives more salary

```python
df_sorted = df.sort_values(by='work_year')

plt.figure(figsize=(10, 6))
sns.lineplot(x=df_sorted['work_year'], y=df_sorted['salary_in_usd'],color = 'r')
plt.title('Trend of Cybersecurity Salaries')
plt.xlabel('Work Year')
plt.ylabel('Salary (USD)')
plt.show()
```

 - Salary of cyber security has been risen after 2021

```python
plt.figure(figsize = (10,6))
sns.barplot(data = df,x='employment_type',y='salary_in_usd',palette = 'Set2')
plt.title('Salary Distribution on  employment type')
plt.show()
```

 - Employment type impacts on salary, fulltime postions offering high compensation

```python
plt.figure(figsize = (10,6))
df.groupby('company_size')['employment_type'].value_counts().plot(kind = 'bar',color ='r')
plt.title('Employment Types by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.xticks(rotation=45) 

plt.show()
```

 - Full-Time employment is the dominant type across all company sizes, while Contract, Part-Time, and Freelance positions are less prevalent

```python
sns.distplot(df,x=df['Price_euros'])
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price (Euros)')
plt.show()
```

 - Most laptops are priced between 1000 and 2000 Euros.

```python
average_prices = df.groupby('Company')['Price_euros'].mean().sort_values(ascending=False)
```

```python
average_prices = df.groupby('Company')['Price_euros'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Laptop Prices by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price (Euros)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
```

 - Razer has the highest average laptop price among the listed brands, followed by LG and MSI.
 - Brands like Acer, Chuwi, Mediacom, and Vero have comparatively lower average laptop prices.
 - Apple's average laptop price is relatively high, reflecting its positioning as a premium brand in the market.

```python
type_group = df.groupby('TypeName')['Price_euros'].mean().sort_values(ascending = False)
```

```python
plt.figure(figsize=(10, 6))
type_group.plot(kind='bar', color='lightgreen')  # Changing the color to light green
plt.title('Average Prices of Laptops by Type', fontsize=16)
plt.xlabel('Laptop Type', fontsize=14)
plt.ylabel('Average Price (euros)', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

 - Workstations have the highest average laptop prices, indicating they are designed for professional and high-performance computing tasks, thus commanding a premium price.
 - Gaming laptops also have a relatively high average price, reflecting the advanced hardware specifications required for gaming purposes.
 - Notebooks have a lower average price compared to other types, making them more accessible to a wider range of consumers.
 - Netbooks have the lowest average price, indicating they are budget-friendly options designed for basic computing needs.





```python
opsys_prices = df.groupby('OpSys')['Price_euros'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
opsys_prices.plot(kind='bar', color='red')
plt.title('Average Laptop Prices by Operating System')
plt.xlabel('Operating System')
plt.ylabel('Average Price (Euros)')
plt.xticks(rotation=45)
plt.show()
```

- Laptops with macOS and Windows 7 have the highest average prices, reaching around 1750 Euros.
- Laptops with the Android operating system are the most affordable, with an average price below 250 Euros
     - The choice of operating system significantly impacts the cost of laptops, with macOS and Windows 7 commanding premium prices while Android-based laptops remain budget-friendly.

```python
ram_price = df.groupby('Ram')['Price_euros'].mean().reset_index()

ram_price = ram_price.sort_values(by='Ram')

plt.figure(figsize=(10, 6))
sns.barplot(x='Ram', y='Price_euros', data=ram_price)
plt.title('Average Price by RAM Size')
plt.xlabel('RAM Size (GB)')
plt.ylabel('Average Price (Euros)')
plt.xticks(rotation=45)
plt.show()
```

 - Laptops with 32GB of RAM are on average the most expensive.
 - The price decreases for laptops with more than 32GB RAM.
 - Laptops with 2GB and 64GB RAM are the cheapest on average.

```python
df['Memory'].unique()
```

```python
SSD = []
HDD = []
Flash_Storage = []
Hybrid = []

for i in df['Memory']:
    if 'SSD' in i:
        SSD.append(i)
    elif 'HDD' in i:
        HDD.append(i)
    elif 'Flash Storage' in i:
        Flash_Storage.append(i)
    elif 'Hybrid' in i:
        Hybrid.append(i)
```

```python
avg_price_ssd = df[df['Memory'].str.contains('SSD')]['Price_euros'].mean()
avg_price_hdd = df[df['Memory'].str.contains('HDD')]['Price_euros'].mean()
avg_price_flash_storage = df[df['Memory'].str.contains('Flash Storage')]['Price_euros'].mean()
avg_price_hybrid = df[df['Memory'].str.contains('Hybrid')]['Price_euros'].mean()

plt.figure(figsize=(10, 6))
sns.barplot(x=['SSD', 'HDD', 'Flash Storage', 'Hybrid'], y=[avg_price_ssd, avg_price_hdd, avg_price_flash_storage, avg_price_hybrid])
plt.title('Average Price by Memory Type')
plt.xlabel('Memory Type')
plt.ylabel('Average Price (Euros)')
plt.show()
```

 - Laptops with SSD  are the most expensive, with an average price around 1400 Euros.
 - HDD  laptops have an average price slightly above 1000 Euros.
 - Laptops with Flash Storage are more affordable, with an average price just above 400 Euros.
 - Hybrid laptops fall in between, with an average price close to 1200 Euros.

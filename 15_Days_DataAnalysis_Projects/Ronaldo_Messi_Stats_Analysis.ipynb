```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Documents\15 days project datasets\Cristinao vs Messi.csv")
df
```

```python
df.info()
```

```python
ronaldo_data = df[df['Player'] == 'Cristiano Ronaldo']['Playing_Position'].mode()[0]
messi_data = df[df['Player'] == 'Lionel Messi']['Playing_Position'].mode()[0]
```

```python
ronaldo_data
```

```python
messi_data
```

```python
df.loc[df['Player'] == 'Cristiano Ronaldo', 'Playing_Position'] = df.loc[df['Player'] == 'Cristiano Ronaldo', 'Playing_Position'].fillna(ronaldo_data)
df.loc[df['Player'] == 'Lionel Messi', 'Playing_Position'] = df.loc[df['Player'] == 'Lionel Messi', 'Playing_Position'].fillna(messi_data)
```

```python
df.isnull().sum()
```

```python
df['Type'].unique()
```

```python
ronaldotype_data = df[df['Player'] == 'Cristiano Ronaldo']['Type'].mode()[0]
messitype_data = df[df['Player'] == 'Lionel Messi']['Type'].mode()[0]


df.loc[df['Player'] == 'Cristiano Ronaldo', 'Type'] = df.loc[df['Player'] == 'Cristiano Ronaldo', 'Type'].fillna(ronaldotype_data)
df.loc[df['Player'] == 'Lionel Messi', 'Type'] = df.loc[df['Player'] == 'Lionel Messi', 'Type'].fillna(messitype_data)
```

```python
total_goals_by_player = df.groupby('Player')['Minute'].count()
total_goals_by_player
```

```python
goal_type_frequency = df.groupby(['Player', 'Type']).size().unstack(fill_value=0)
goal_type_frequency
```

```python
plt.figure(figsize=(10, 6))
goal_type_frequency.loc['Cristiano Ronaldo'].plot(kind='bar', color='blue')
plt.title('Goal Type Frequency for Cristiano Ronaldo')
plt.xlabel('Goal Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(10, 6))
goal_type_frequency.loc['Lionel Messi'].plot(kind='bar', color='red')
plt.title('Goal Type Frequency for Lionel Messi')
plt.xlabel('Goal Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

 - Cristiano Ronaldo has scored more goals overall compared to Lionel Messi.
 - Both players have a significant number of goals from right-footed shots, but Messi has more goals from left-footed shots.
 - Messi has a higher number of goals from direct free kicks compared to Ronaldo.
 - Ronaldo has scored more headers than Messi.
 - Messi has scored more goals from solo runs compared to Ronaldo.

```python
seasonal_goals = df.groupby(['Season', 'Player']).size().unstack(fill_value=0)
seasonal_goals
```

```python
plt.figure(figsize=(12, 6))
seasonal_goals.plot(kind='line', marker='o')
plt.title('Seasonal Trends in Goal Scoring')
plt.xlabel('Season')
plt.ylabel('Total Goals')
plt.xticks(rotation=45)
plt.legend(title='Player', loc='upper left')
plt.grid(True)
plt.show()
```

 - Ronaldo's peak seasons were 07/08, 10/11, and 14/15, where he scored 42, 53, and 61 goals respectively.
 - Messi's most prolific seasons were 11/12, 14/15, and 16/17, with 73, 58, and 54 goals respectively.
 - The 11/12 season was exceptional for Messi, with a record-breaking 73 goals, while Ronaldo's peak was in the 14/15 season with 61 goals.

```python
goals_by_competition = df.groupby(['Competition', 'Player']).size().unstack(fill_value=0)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(goals_by_competition, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Goals Scored by Players in Different Competitions')
plt.xlabel('Player')
plt.ylabel('Competition')
plt.show()
```

 - Cristiano Ronaldo scored the most goals in LaLiga, reaching 311, while Lionel Messi surpassed him with 474 goals in the same league.
 - Both players excelled in the UEFA Champions League, Ronaldo with 140 goals and Messi with 128.
 - Messi outperformed Ronaldo in the Copa del Rey, scoring 56 goals compared to Ronaldo's 22.
 - Ronaldo showcased his talent in the Premier League, scoring 103 goals, while Messi has not scored in this league.
 - Messi's remarkable performance in the Supercopa is evident, scoring 14 goals, while Ronaldo scored 4.
 - Ronaldo contributed significantly in Serie A with 81 goals, a competition where Messi has not scored yet.

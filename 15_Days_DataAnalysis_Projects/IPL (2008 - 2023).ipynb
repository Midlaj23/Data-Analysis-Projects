```python
super_over_count = df['super_over'].sum()

super_over_wins = df[df['super_over'] == True]['winner'].value_counts()
total_super_overs = super_over_wins.sum()
super_over_success_rate = (super_over_wins / total_super_overs) * 100

print("Number of matches that went into Super Overs:", super_over_count)
print("Success rates of teams in Super Overs:")
print(super_over_success_rate)
```

 - This indicates that SRH has been the most successful team in Super Overs, while DC, KKR, and RCB have the lowest success rates among the teams analyzed.

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='season', data=df, palette='inferno')
plt.title('Distribution of Matches Across Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
```

 - The number of matches peaked in 2013 and 2023  with over 70 matches played.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='winner', data=df, palette='viridis', order=df['winner'].value_counts().index)
plt.title('Winning Teams')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=90)
plt.show()
```

 - Mumbai Indians have the highest number of wins, exceeding 120 victories.
 -  Chennai Super Kings (CSK), Kolkata Knight Riders (KKR), Royal Challengers Bangalore (RCB), and Sunrisers Hyderabad (SRH) follow with significant wins but less than MI
 -  Teams like Gujarat Lions (GL) and Kochi have the least number of wins.

```python
plt.figure(figsize=(8, 8))
decision_counts = df['decision'].value_counts()
plt.pie(decision_counts, labels=decision_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Toss Decision Distribution')
plt.show()
```

 - Teams choose to bowl first in 64.1% of cases.
 - Teams opt to bat first in 35.9% of cases.

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='venue_name', data=df, palette='Set2', order=df['venue_name'].value_counts().head(10).index)
plt.title('Top 10 Venues with Most Matches')
plt.xlabel('Number of Matches')
plt.ylabel('Venue')
plt.show()
```

 - Wankhede Stadium, Mumbai :  Hosted the highest number of matches.
 -  M. Chinnaswamy Stadium, Bengaluru: Also among the top venues

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='pom', data=df, palette='rainbow', order=df['pom'].value_counts().head(10).index)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=90)
plt.show()
```

 - Top players with the most Player of the Match awards are AB de Villiers, Chris Gayle, and Rohit Sharma.
